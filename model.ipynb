```python
#import the dependencies

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

import pickle
```

```python
# data collection - kaggle
wine = pd.read_csv('winequality-red.csv')
display(wine.head())
print('no of rows and columns: ',wine.shape)
```

```python
wine.isnull().sum()
```

```python
#statistical measure
wine.describe()
```

```python
#number of values for each quality
sns.catplot(x='quality', data = wine, kind='count')
```

```python
#All Col vs  quality
col =['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol']	
for col in col:
 plot = plt.figure(figsize=(5,5))
 sns.barplot(x='quality',y=col,data=wine)
```

```python
# correlation 
#1. Positive correlation
#2. Negative correlation

correlation = wine.corr()

# constructiong a heatmap to understand the correlation of the columns

plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot = True, annot_kws={'size':8}, cmap= 'Greens')
```

```python
#seperate the data and label

X = wine.drop('quality', axis=1)


display(X)
```

```python
#Label Binarization or label Encoding as good and bad
Y = wine['quality'].apply(lambda y_value: 1 if y_value >= 7 else 0)
print(Y)
```

```python
#Train and Test split of Wine data

X_train,X_test,Y_train,Y_test   =  train_test_split(X,Y,test_size=0.2,random_state = 3)
```

```python
print(Y.shape,Y_train.shape,Y_test.shape)
print(X.shape,X_train.shape,X_test.shape)
```

```python
#Model Training: Random Forest Classifier

model = RandomForestClassifier()
model.fit(X_train,Y_train)
```

```python
#model evaluation

#accuracy_score on test data

X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)

display('Accuracy of test data: ', test_data_accuracy)
```

```python
#Buid a predictive system
import numpy as np


input_data = (7.4,0.7,0.0,1.9,0.076,11.0,34.0,0.9978,3.51,0.56,9.4)

# changing the input data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshape the data as we are the predicting the label for only one instance

input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if(prediction[0]==1):
 print('Good Quality Wine')
else:
 print('Bad Quality Wine')
```

```python
#save the trained model


filename='trained_model.sav'
pickle.dump(model,open(filename,'wb')) #wb= writebinary
```
